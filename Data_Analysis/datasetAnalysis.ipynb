```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

dataset = pd.read_csv("final_dataset.csv")

count = 1

counters = {'FUNCIONALIDADE': 0, 'HARDWARE': 0, 'SOFTWARE': 0, 'COMUNICAO': 0, 'PESSOA': 0}
all_partitions = []

for data in dataset.values:
    current_count = data[5]

    if current_count != count:
        plt.figure(figsize=(10, 6))
        plt.barh(list(counters.keys()), list(counters.values()), color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'])
        plt.title(f'Entidades por Classe para o Grupo {count}')
        plt.xlabel('Quantidade de Entidades')
        plt.ylabel('Classes')
        plt.show()
        
        all_partitions.append(list(counters.values()))
        counters = {key: 0 for key in counters}
        count += 1

    if data[2] != 'O':
        if 'FUNCIONALIDADE' in data[2]: 
            counters['FUNCIONALIDADE'] += 1
        elif 'HARDWARE' in data[2]: 
            counters['HARDWARE'] += 1
        elif 'SOFTWARE' in data[2]: 
            counters['SOFTWARE'] += 1
        elif 'COMUNICAO' in data[2]: 
            counters['COMUNICAO'] += 1
        elif 'PESSOA' in data[2]: 
            counters['PESSOA'] += 1

if sum(counters.values()) > 0:
    plt.figure(figsize=(10, 6))
    plt.barh(list(counters.keys()), list(counters.values()), color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'])
    plt.title(f'Entidades por Classe para o Grupo {count}')
    plt.xlabel('Quantidade de Entidades')
    plt.ylabel('Classes')
    plt.show()

    all_partitions.append(list(counters.values()))

```

```python
average = np.mean(all_partitions, axis=0)
print(average)
```
